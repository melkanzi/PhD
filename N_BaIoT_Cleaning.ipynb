## Data Preprocessing Pipeline for N-BaIoT Dataset

This notebook will guide you through the preprocessing steps for the N-BaIoT dataset, which is assumed to be in a zip file on your Google Drive.

### Step 1: Mount Google Drive and Unzip the Dataset

First, we need to mount Google Drive to access your files. After mounting, please update the `zip_file_path` variable with the correct path to your N-BaIoT zip file on Drive.

```python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Define the path to your zip file on Google Drive
# IMPORTANT: Update this path to your specific file location
zip_file_path = '/content/drive/MyDrive/Datasets/N-BaIoT.zip' # Example path, please adjust!

# Define the directory where you want to extract the files
extraction_path = './N-BaIoT_extracted'

import os
import zipfile

# Create the extraction directory if it doesn't exist
os.makedirs(extraction_path, exist_ok=True)

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_path)

print(f"Dataset unzipped to: {extraction_path}")

# List the contents of the extracted folder to identify the data files
print("Contents of extracted folder:")
for root, dirs, files in os.walk(extraction_path):
    for name in files:
        print(os.path.join(root, name))
```

### Step 2: Load the Dataset

Assuming the unzipped dataset contains CSV files, we will load them into a Pandas DataFrame. You might need to adjust the `data_file_path` to point to the actual data file within the extracted directory.

```python
import pandas as pd
import os

all_dfs = []

# Iterate through each item in the extraction path
for item_name in os.listdir(extraction_path):
    item_path = os.path.join(extraction_path, item_name)

    # Check if it's a directory (assumed to be a device folder)
    if os.path.isdir(item_path):
        benign_file_path = os.path.join(item_path, 'benign_traffic.csv')

        if os.path.exists(benign_file_path):
            try:
                temp_df = pd.read_csv(benign_file_path)
                # Add columns to identify the device and traffic type
                temp_df['device_category'] = item_name  # Device name from folder name
                temp_df['traffic_type'] = 'benign' # This file contains benign traffic
                all_dfs.append(temp_df)
                print(f"Loaded {benign_file_path} with shape {temp_df.shape}")
            except Exception as e:
                print(f"Error loading {benign_file_path}: {e}")

if all_dfs:
    df = pd.concat(all_dfs, ignore_index=True)
    print("\nAll benign traffic datasets loaded and concatenated successfully.")
    print(f"Final shape of the dataset: {df.shape}")
    display(df.head())
else:
    print("No 'benign_traffic.csv' files were found or loaded in the subdirectories.")
    df = pd.DataFrame() # Initialize an empty DataFrame if nothing was loaded
```

### Step 3: Handle Missing Values

We will remove rows with any missing values from the dataset. This is a simple approach; more sophisticated methods (like imputation) could be used if necessary.

```python
print(f"Original shape: {df.shape}")
missing_values_count = df.isnull().sum().sum()

if missing_values_count > 0:
    df_cleaned_missing = df.dropna()
    print(f"Removed {df.shape[0] - df_cleaned_missing.shape[0]} rows with missing values.")
    print(f"New shape after handling missing values: {df_cleaned_missing.shape}")
    df = df_cleaned_missing
else:
    print("No missing values found.")

display(df.head())
```

### Step 4: Remove Duplicate Rows

Identify and remove any exact duplicate rows in the dataset.

```python
print(f"Shape before removing duplicates: {df.shape}")
duplicates_count = df.duplicated().sum()

if duplicates_count > 0:
    df_no_duplicates = df.drop_duplicates()
    print(f"Removed {duplicates_count} duplicate rows.")
    print(f"Shape after removing duplicates: {df_no_duplicates.shape}")
    df = df_no_duplicates
else:
    print("No duplicate rows found.")

display(df.head())
```

### Step 5: Fix Inconsistent Formats (Initial Check)

This step typically involves inspecting column data types and unique values to identify inconsistencies. Without specific knowledge of the N-BaIoT dataset's columns, we'll perform a general check and conversion for common types. You might need to customize this based on your data.

We'll ensure numeric columns are indeed numeric and handle potential errors during conversion.

```python
print("Data types before format fixing:")
print(df.info())

# Columns identified as categorical labels that should NOT be converted to numeric
# These will be handled by One-Hot Encoding later.
categorical_label_cols = ['device_category', 'traffic_type']

# Attempt to convert all columns that should be numeric to numeric, coercing errors
for col in df.columns:
    if col in categorical_label_cols:
        print(f"Skipping numerical conversion for identified categorical label column: '{col}'")
        continue # Skip to the next column

    if df[col].dtype == 'object':
        try:
            # Try converting to numeric, if it fails, keep as object or convert to category
            original_col_data = df[col].copy() # Keep a copy to compare after conversion
            df[col] = pd.to_numeric(df[col], errors='coerce')

            # If conversion was successful, check for new NaNs introduced by coercion
            # These new NaNs mean there were non-numeric strings in a column expected to be numeric.
            newly_nan_rows = df[df[col].isnull() & original_col_data.notnull()].shape[0]
            if newly_nan_rows > 0:
                initial_rows = df.shape[0]
                df.dropna(subset=[col], inplace=True)
                print(f"Removed {initial_rows - df.shape[0]} rows due to non-numeric values in '{col}' after attempted conversion.")
        except Exception:
            # If conversion fails for other reasons, it's likely a true string/categorical column
            # but not in our explicit skip list, so we might want to inspect it.
            print(f"Warning: Could not convert column '{col}' to numeric, keeping as original type.")

print("\nData types after initial format fixing:")
print(df.info())
display(df.head())
```

### Step 6: Convert Text Labels to Numeric Form (One-Hot Encoding)

Assuming there is a 'device_category' column that contains text labels, we will apply One-Hot Encoding to convert it into a numeric format. If your categorical column has a different name, please adjust `categorical_column`.

```python
from sklearn.preprocessing import OneHotEncoder

# IMPORTANT: Identify your categorical column for device categories
# For N-BaIoT, this might be 'device_category', 'type', or a similar label column.
# Inspect df.columns and df['your_label_column'].unique() if unsure.

categorical_column = 'device_category' # Placeholder, adjust to your actual column name

if categorical_column in df.columns and df[categorical_column].dtype == 'object':
    print(f"Applying One-Hot Encoding to '{categorical_column}'...")
    encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
    encoded_features = encoder.fit_transform(df[[categorical_column]])

    # Create a DataFrame from the encoded features
    encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out([categorical_column]), index=df.index)

    # Drop the original categorical column and concatenate the new encoded columns
    df = pd.concat([df.drop(columns=[categorical_column]), encoded_df], axis=1)
    print("One-Hot Encoding applied.")
    print(f"Shape after One-Hot Encoding: {df.shape}")
else:
    print(f"Categorical column '{categorical_column}' not found or is not of object dtype. Skipping One-Hot Encoding.")

display(df.head())
```

### Step 7: Apply Normalization (Min-Max Scaling)

We will apply Min-Max scaling to all numerical features to bring them into a range of [0, 1]. This helps in improving the performance of many machine learning algorithms.

```python
from sklearn.preprocessing import MinMaxScaler

print("Applying Min-Max Normalization...")

# Identify numerical columns for scaling
# Exclude any one-hot encoded columns if you want to scale only original numerical features
# For simplicity, we'll scale all numerical columns here.
numerical_cols = df.select_dtypes(include=['number']).columns

if len(numerical_cols) > 0:
    scaler = MinMaxScaler()
    df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
    print("Min-Max Normalization applied to numerical columns.")
else:
    print("No numerical columns found to apply Min-Max Normalization.")

display(df.head())
```

### Step 8: Split the Data (Stratified Random Split)

We will perform a 70%-30% stratified random split based on the 'device_category' (or the one-hot encoded representation of it) to ensure that each device type is adequately represented in both the training and test sets. You will need to specify your target column for stratification.

```python
from sklearn.model_selection import train_test_split
import numpy as np

# Identify columns related to device categories after One-Hot Encoding
target_categorical_prefix = 'device_category_'
ohe_device_category_cols = [col for col in df.columns if col.startswith(target_categorical_prefix)]

# Also identify 'traffic_type' if it's present and should be treated as a label/excluded from features
# Assuming 'traffic_type' is a label and not a feature for the main classification task.
other_label_cols = ['traffic_type']

if not ohe_device_category_cols:
    print("Error: No One-Hot Encoded device category columns found. Cannot perform stratified split.")
    print("Please ensure 'device_category' was correctly one-hot encoded in the previous step.")
    X_train, X_test, y_train, y_test = None, None, None, None
else:
    # Reconstruct a single categorical series for stratification purposes
    # This assumes that for each row, exactly one of the ohe_device_category_cols is 1.
    y_stratify = df[ohe_device_category_cols].idxmax(axis=1)

    # Define features (X) by dropping all label-related columns
    features_to_drop = ohe_device_category_cols + other_label_cols
    X = df.drop(columns=features_to_drop, errors='ignore')

    # Define the actual labels (y) as the one-hot encoded device categories
    y = df[ohe_device_category_cols]

    # Perform the stratified split using y_stratify for stratification
    X_train, X_test, y_train_ohe, y_test_ohe, y_stratify_train, y_stratify_test = train_test_split(
        X, y, y_stratify, test_size=0.3, random_state=42, stratify=y_stratify
    )

    # For the final export, y_train and y_test should be the one-hot encoded labels
    y_train = y_train_ohe
    y_test = y_test_ohe

    print("Data split into training and testing sets (70%-30% stratified).")
    print(f"X_train shape: {X_train.shape}")
    print(f"X_test shape: {X_test.shape}")
    print(f"y_train (OHE labels) shape: {y_train.shape}")
    print(f"y_test (OHE labels) shape: {y_test.shape}")

    # You can also check the distribution of classes in train and test sets using the stratification target
    print("\nClass distribution in original dataset (stratification target):")
    print(y_stratify.value_counts(normalize=True))
    print("\nClass distribution in training set (stratification target):")
    print(y_stratify_train.value_counts(normalize=True))
    print("\nClass distribution in test set (stratification target):")
    print(y_stratify_test.value_counts(normalize=True))

    # train_set and test_set variables are not strictly needed for the final .npy export
    # but if needed, they would be constructed like this:
    # train_set = pd.concat([X_train, y_train], axis=1)
    # test_set = pd.concat([X_test, y_test], axis=1)
```

### Step 9: Export Training and Test Sets as .npy Files

Finally, we will save the preprocessed training and test sets as NumPy `.npy` files for later use.

```python
if X_train is not None and X_test is not None and y_train is not None and y_test is not None:
    output_dir = './preprocessed_N-BaIoT'
    os.makedirs(output_dir, exist_ok=True)

    # Save features and labels separately for flexibility
    np.save(os.path.join(output_dir, 'X_train.npy'), X_train.to_numpy())
    np.save(os.path.join(output_dir, 'X_test.npy'), X_test.to_numpy())
    np.save(os.path.join(output_dir, 'y_train.npy'), y_train.to_numpy())
    np.save(os.path.join(output_dir, 'y_test.npy'), y_test.to_numpy())

    print(f"Preprocessed data saved to '{output_dir}' as .npy files:")
    print("  X_train.npy")
    print("  X_test.npy")
    print("  y_train.npy")
    print("  y_test.npy")

else:
    print("Skipping .npy export as data splitting was not successful.")
```
